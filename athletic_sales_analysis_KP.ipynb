```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
athletic_sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
blended_sales_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df]).reset_index(drop=True)
blended_sales_df
```

```python
# Check if any values are null.
null_values = blended_sales_df.isnull().sum()
null_values
```

```python
# Check the data type of each column
blended_sales_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
blended_sales_df["invoice_date"] = pd.to_datetime(blended_sales_df['invoice_date'],errors='coerce')
```

```python
# Confirm that the "invoice_date" data type has been changed.
blended_sales_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_product_sales = blended_sales_df.groupby(['region', 'state', 'city'])['units_sold'].agg(Total_Product_Sales= ('sum'))

# Show the top 5 results.
total_product_sales.sort_values(by=(["Total_Product_Sales"]), ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_product_sales = pd.pivot_table(blended_sales_df, index=(['region', 'state', 'city']), values= 'units_sold', aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_product_sales = total_product_sales.rename(columns={"units_sold": "Total_Product_Sales"})

# Show the top 5 results.
total_product_sales.sort_values(by=(["Total_Product_Sales"]), ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = blended_sales_df.groupby(['region', 'state', 'city'])['total_sales'].agg(Total_Sales= ('sum'))


# Show the top 5 results.
total_sales.sort_values(by=(["Total_Sales"]), ascending=False).head(5)
```
